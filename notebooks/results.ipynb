```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('../db/results.db')  # Replace 'your_database.db' with your actual database file

# Write your SQL query
query = 'SELECT * FROM RESULTS'  # Replace 'your_table' with the actual table name

# Use pandas to read the SQL query result into a DataFrame
df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

df.head(60)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming df is your DataFrame
# Create a grid of subplots for each unique operation with three parallel columns
unique_operations = df[df.operation != 'example_operation']['operation'].unique()
unique_operations = [operation for operation in unique_operations if operation is not None]
num_operations = len(unique_operations)

# Calculate the number of rows needed for three columns
num_rows = (num_operations + 2) // 3

# Set up the figure and axes with increased lateral separation
fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 6 * num_rows))
fig.suptitle('CC AUC and PCE AUC vs Parameter for Each Operation', fontsize=16)

# Flatten the axes array to simplify indexing
axes = axes.flatten()

# Iterate through each unique operation and plot the data
for i, operation in enumerate(unique_operations):
    # Filter the DataFrame for the current operation
    operation_df = df[df['operation'] == operation]

    # Sort the DataFrame by the 'parameter' column in ascending order
    operation_df = operation_df.sort_values(by='parameter')

    # Set the positions for the bars
    positions = np.arange(len(operation_df['parameter']))

    # Plot CC AUC and PCE AUC as dual bars
    axes[i].bar(positions - 0.2, operation_df['auc_cc'], width=0.35, label='CC AUC')
    axes[i].bar(positions + 0.2, operation_df['auc_pce'], width=0.35, label='PCE AUC', color='orange')

    axes[i].set_title(f'{operation} - CC AUC and PCE AUC vs Parameter')
    axes[i].set_xlabel('Parameter')
    axes[i].set_ylabel('AUC Values')
    
    # Round the xticks to two decimal places and rotate 45 degrees
    axes[i].set_xticks(positions)
    axes[i].set_xticklabels([f'{val:.2f}' for val in operation_df['parameter']])
    axes[i].tick_params(axis='x', rotation=45)  # Adjusted rotation value
    
    # Set x-axis limits
    axes[i].set_xlim(0.4, len(positions) - 0.5)
    axes[i].set_ylim(0.75, 1)
    
    axes[i].legend(loc='lower center', bbox_to_anchor=(0.5, 0), fancybox=True, shadow=True, ncol=2)

# Hide empty subplots if any
for i in range(num_operations, len(axes)):
    axes[i].axis('off')

# Adjust layout and show the plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('../documentation/report/images/experiment_5.png')
plt.show()





```

```python
def labelTest(devices):
    if devices == 'D44,D45,D46':
        return 1e-3, 1e3, 'med' #med
    elif devices == 'D47,D48,D49':
        return 0.9, 1e4, 'high' #high
    elif devices == 'D41,D42,D43':
        return 1e-9, 1e11, 'low' #low
    elif devices == 'D51,D52,D53':
        return 1e-6, 1e9, 'low3' #low3
    elif devices == 'D54,D55,D56':
        return 1e-7, 1e10, 'low4' #low4


df3 = df[df['experiment'].isin([7, 6])].sort_values(by=['operation', 'parameter'], ascending=[False, True])

df3['content_weight'] = df3['deviceList'].apply(lambda x: labelTest(x)[0])
df3['style_weight'] = df3['deviceList'].apply(lambda x: labelTest(x)[1])
df3['method'] = df3['deviceList'].apply(lambda x: labelTest(x)[2])

columns = ['wl_id', 'experiment', 'auc_cc', 'auc_pce', 'content_weight', 'style_weight', 'method']
df3 = df3[columns].drop_duplicates(subset=['content_weight', 'style_weight'])
#df3['style_strength'] = []

#df3.head(50)
df3.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df3 is your DataFrame
methods = df3['method'].unique()
bar_width = 0.35  # Width of the bars
gap_width = 0   # Gap width between sets of dual barplots

fig, ax = plt.subplots(figsize=(15, 12))

# Combine method_data for each method into a single list
bars = []

for i, method in enumerate(methods):
    method_data = df3[df3['method'] == method]

    # Bar positions
    index = np.arange(len(method_data)) * (bar_width * 2 + gap_width) + i * (bar_width * 2 + gap_width)

    # Plot auc_cc
    bar1 = ax.bar(index, method_data['auc_cc'], bar_width, label=f'AUC_CC ({method})', color='blue')

    # Plot auc_pce
    bar2 = ax.bar(index + bar_width + gap_width, method_data['auc_pce'], bar_width, label=f'AUC_PCE ({method})', color='orange')

    # Combine the bars for each method into a single list
    bars.extend([bar1, bar2])

# Set labels, title, and legend
#ax.set_xlabel('Style and Content weights in sytle transfer transformations', labelpad=10)  # Adjust the distance of the x-axis label from the plot
ax.set_ylabel('AUC')
ax.set_title('AUC values in tests using Style Transfer transformations', fontsize=18)  # Increase title font size

# Format style_weight and content_weight in scientific notation and round to max two decimal places
formatted_labels = [
    f"Content Weight: {cw:.2e}\nStyle Weight: {sw:.2e}" for sw, cw in zip(df3['style_weight'], df3['content_weight'])
]
ax.set_xticks(np.arange(len(df3)) * (bar_width * 2 + gap_width) + bar_width / 2)
ax.set_xticklabels(formatted_labels, rotation=25, ha='right')

# Remove legend
ax.legend(labels=['CC AUC', 'PCE AUC'])

# Add labels on top of each bar
for bar in bars:
    for rect in bar:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
plt.savefig('../documentation/report/images/style_transfer_auc.png')
plt.show()
```
